```python
import numpy as np
from submit import my_map
from submit import my_fit
from submit import my_decode
import time as tm
```

```python
Z_trn = np.loadtxt( "secret_trn.txt" )
Z_tst = np.loadtxt( "secret_tst.txt" )

n_trials = 5

d_size = 0
t_train = 0
t_map = 0
acc = 0
```

```python
for t in range( n_trials ):
  tic = tm.perf_counter()
  w, b = my_fit( Z_trn[:, :-1], Z_trn[:,-1] )
  toc = tm.perf_counter()

  t_train += toc - tic
  w = w.reshape( -1 )

  d_size += w.shape[0]

  tic = tm.perf_counter()
  feat = my_map( Z_tst[:, :-1] )
  toc = tm.perf_counter()
  t_map += toc - tic

  scores = feat.dot( w ) + b

  pred = np.zeros_like( scores )
  pred[ scores > 0 ] = 1

  acc += np.average( Z_tst[ :, -1 ] == pred )
```

```python
d_size /= n_trials
t_train /= n_trials
t_map /= n_trials
acc /= n_trials
```

```python
def get_model( p, q, r, s ):
  p = np.maximum( p, 0 )
  q = np.maximum( q, 0 )
  r = np.maximum( r, 0 )
  s = np.maximum( s, 0 )
  d = p - q
  c = r - s
  alpha = ( d + c ) / 2
  beta = ( d - c ) / 2
  w = np.zeros( ( len( alpha ) + 1, )  )
  w[:-1] += alpha
  w[1:] += beta
  return w
```

```python
W = np.loadtxt( "secret_mod.txt" )
( n_models, dims ) = W.shape
t_decode = 0
m_dist = 0
for t in range( n_trials ):
  for itr in range( n_models ):
    w = W[ itr, : ]
    tic = tm.perf_counter()
    p_hat, q_hat, r_hat, s_hat = my_decode( w )
    toc = tm.perf_counter()
    t_decode += toc - tic
    w_hat = get_model( p_hat, q_hat, r_hat, s_hat )
    m_dist += np.linalg.norm( w - w_hat )
```

```python
t_decode /= ( n_trials * n_models )
m_dist /= ( n_trials * n_models )
```

```python
print( f"d_size :{d_size}, t_train : {t_train}, t_map :{t_map}, 1-acc: {1 - acc}, t_decode : {t_decode}, m_dist : {m_dist}" )
```

```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from submit import my_map, my_fit
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression

# Assume Z_trn and Z_tst are loaded: last column is labels
X_trn, y_trn = Z_trn[:, :-1], Z_trn[:, -1]
X_tst, y_tst = Z_tst[:, :-1], Z_tst[:, -1]

results = []

# Hyperparameter grids
svc_params = {
    'loss': ['hinge', 'squared_hinge'],
    'C': [0.01, 1, 100]
}
log_params = {
    'C': [0.01, 1, 100]
}

# Evaluate LinearSVC
for loss in svc_params['loss']:
    for C in svc_params['C']:
        # Train and time
        start = time.perf_counter()
        model = LinearSVC(loss=loss, C=C, max_iter=10000)
        model.fit(my_map(X_trn), y_trn)
        train_time = time.perf_counter() - start
        
        # Test accuracy
        feat_tst = my_map(X_tst)
        preds = model.predict(feat_tst)
        accuracy = np.mean(preds == y_tst)
        
        results.append({
            'Model': 'LinearSVC',
            'Loss': loss,
            'C': C,
            'TrainTime_s': train_time,
            'Accuracy': accuracy
        })

# Evaluate LogisticRegression
for C in log_params['C']:
    start = time.perf_counter()
    model = LogisticRegression(C=C, max_iter=10000, solver='liblinear')
    model.fit(my_map(X_trn), y_trn)
    train_time = time.perf_counter() - start
    
    feat_tst = my_map(X_tst)
    preds = model.predict(feat_tst)
    accuracy = np.mean(preds == y_tst)
    
    results.append({
        'Model': 'LogisticRegression',
        'Loss': 'log-loss',
        'C': C,
        'TrainTime_s': train_time,
        'Accuracy': accuracy
    })

# Create DataFrame
df_results = pd.DataFrame(results)

# Display the results table
print("\nHyperparameter Experiment Results:\n")
print(df_results.to_string(index=False))

# Plot Accuracy vs C
plt.figure(figsize=(8, 5))
for model in df_results['Model'].unique():
    for loss in df_results[df_results['Model'] == model]['Loss'].unique():
        subset = df_results[(df_results['Model'] == model) & (df_results['Loss'] == loss)]
        plt.plot(subset['C'], subset['Accuracy'], marker='o', label=f"{model} ({loss})")
plt.xscale('log')
plt.xlabel('C (log scale)')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs C')
plt.legend()
plt.grid(True)
plt.show()

# Plot Training Time vs C
plt.figure(figsize=(8, 5))
for model in df_results['Model'].unique():
    for loss in df_results[df_results['Model'] == model]['Loss'].unique():
        subset = df_results[(df_results['Model'] == model) & (df_results['Loss'] == loss)]
        plt.plot(subset['C'], subset['TrainTime_s'], marker='o', label=f"{model} ({loss})")
plt.xscale('log')
plt.xlabel('C (log scale)')
plt.ylabel('Training Time (s)')
plt.title('Training Time vs C')
plt.legend()
plt.grid(True)
plt.show()
```
